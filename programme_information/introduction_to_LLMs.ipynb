# Introduction to LLMs

## Course description

This course is designed to enhance your understanding of modern Large Language Models (LLMs) and how they work. Taking a hands-on approach, it combines conceptual slides with interactive Jupyter notebooks to guide you from the fundamentals of machine learning and neural networks through to the architecture underlying today's state-of-the-art models. You will finish the session equipped with practical skills for working with LLMs effectively, including how to interpret tokenization, understand semantic embeddings, and craft effective prompts.


```python
import pandas as pd
from datetime import datetime
from html import escape
from IPython.display import display, HTML

# Define the course that is being looked at
course_name = "Introduction to Large Language Models"

# Load the CSV file
file_path = '../data/workshop_info.csv'
courses_df = pd.read_csv(file_path)
style_file_path = '../data/course_style.csv'
course_styles_df = pd.read_csv(style_file_path)

# Strip any extra spaces in the column names
courses_df.columns = courses_df.columns.str.strip()
course_styles_df.columns = course_styles_df.columns.str.strip()

# Build a lookup from course name to course style
style_map = {}
if {'course_name', 'course_style'}.issubset(course_styles_df.columns):
    course_styles_df['course_name'] = course_styles_df['course_name'].fillna('').astype(str).str.strip()
    course_styles_df['course_style'] = course_styles_df['course_style'].fillna('').astype(str).str.strip()
    style_map = dict(zip(course_styles_df['course_name'], course_styles_df['course_style']))

# Convert date columns to datetime
courses_df['Start Date'] = pd.to_datetime(courses_df['Start Date'], dayfirst=True, errors='coerce')
courses_df['End Date'] = pd.to_datetime(courses_df['End Date'], dayfirst=True, errors='coerce')

# Get today's date
today = datetime.now()

# Color styles for course-style badges
badge_styles = {
    'Live Coding': ('#0747a6', '#e9f2ff'),
    'Seminar': ('#7a4f01', '#fff2db'),
    'Hybrid': ('#5d2d86', '#f4ebff'),
    'Surgery': ('#0f766e', '#e6fffa'),
    'Self-Paced': ('#1f6f43', '#e8f8ef'),
}

# Function to generate html text based on the course dates and style
def generate_html(row):
    html_parts = []

    if pd.notna(row['Start Date']) and pd.notna(row['End Date']):
        if row['Start Date'] <= today <= row['End Date']:
            html_parts.append("<div style='font-weight: bold; margin-bottom: 0.5rem;'>This course is currently accepting applications.</div>")

    course_style = style_map.get(str(row['Course Name']).strip(), '')
    if isinstance(course_style, str) and course_style.strip():
        style_label = course_style.strip()
        text_color, bg_color = badge_styles.get(style_label, ('#3f3f46', '#f4f4f5'))
        html_parts.append(
            f"<a href='../cfrr_program_details/course_styles.html' title='Read about course delivery styles' style='text-decoration: none; color: inherit; display: block;'>"
            f"<div style='margin-top: 1.2rem; margin-bottom: 1.5rem; padding: 0.82rem 1.12rem; border-left: 8px solid {text_color}; background: #f8fafc; border-radius: 0.52rem; cursor: pointer;'>"
            "<span style='font-size: 1.58rem; font-weight: 800; margin-right: 0.75rem; letter-spacing: 0.01em;'>Course style:</span>"
            f"<span style='display: inline-block; padding: 0.57rem 1.42rem; border-radius: 999px; border: 3px solid {text_color}; color: {text_color}; background-color: {bg_color}; font-size: 1.5rem; font-weight: 800; line-height: 1.1; box-shadow: 0 3px 12px rgba(15, 23, 42, 0.16);'>{escape(style_label)}</span>"
            "</div>"
            "</a>"
        )

    return ''.join(html_parts)

# Apply the function and create a new column for HTML
courses_df['HTML'] = courses_df.apply(generate_html, axis=1)

# Filter the DataFrame for the given course name and display the HTML text
html_output = courses_df[courses_df['Course Name'] == course_name]['HTML'].values[0]
display(HTML(html_output))
```

## Course objectives

* Describe what an LLM does at a high level and explain the concept of next-token prediction.
* Trace the key developments in machine learning and neural networks that led to modern LLMs.
* Explain the core components of a transformer-based LLM: tokenization, embeddings, and attention.
* Understand important practical concepts such as statelessness, context windows, and model parameters.
* Explore how text is tokenized and why token counts matter for cost and context limits.
* Visualise and interpret text embeddings as representations of semantic meaning.
* Apply prompt engineering techniques — including few-shot learning and chain-of-thought reasoning — to improve model outputs.
* Control LLM output behaviour by tuning temperature and sampling parameters.
