# Frequently Asked Questions (FAQ)

**Q1: Who can participate in CfRR workshops?**  
A: Any University of Exeter staff or student (all colleges and campuses welcome). Materials are also openly available for self-study by anyone interested.

**Q2: What topics are available?**  
A: Python, R, Julia, Unix/Linux, HPC, Markdown, version control with Git/GitHub, machine learning, computational thinking, software best practices, and more.

**Q3: How do I find course prerequisites?**  
A: Each module homepage lists prerequisites for its workshop. See the module's main page for details.

**Q4: Where do I access all the course materials?**  
A:  
- [CfRR Jupyter Book website](https://coding-for-reproducible-research.github.io/CfRR_Courses/)
- [CfRR GitHub repository](https://github.com/coding-for-reproducible-research/CfRR_Courses)

**Q5: Who organizes CfRR?**  
A: The programme is run by a dedicated team at the University of Exeter. See our [About Us page](../cfrr_program_details/about_us.md) for the current team, or email us with any questions.

**Q6: How can I contribute or get involved?**  
A: See our [Contributing Guide](../contributing/contributing.ipynb) for roles, community guidelines, and instructions on adding or improving content.

**Q7: Is there a code of conduct?**  
A: Yes! See our [Code of Conduct](../cfrr_program_details/code_of_conduct.ipynb). All participants and contributors are expected to follow it.

**Q8: How do I get help or give feedback?**  
A: See the **Support & Contacts** section below. We welcome all questions, suggestions, and accessibility requests.
